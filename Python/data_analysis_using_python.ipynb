```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'df' is your DataFrame

# Extract data
countries = df['Country or region'][0:10]
scores = df['Score'][0:10]
gdps = df['GDP per capita'][0:10]

# Create a colormap based on GDP values
cmap = plt.cm.get_cmap('viridis')
colors = cmap(gdps / gdps.max())

fig, ax1 = plt.subplots(figsize=(12, 8))  # Create a figure and primary axes

# Plot Score vs Country on the primary axes
ax1.scatter(countries, scores, c=colors, marker='o', s=100, label='Score')
ax1.set_xlabel('Country')
ax1.set_ylabel('Score', color='blue')
ax1.tick_params('y', labelcolor='blue')

# Create a secondary y-axis for GDP
ax2 = ax1.twinx()  # Share the x-axis
ax2.plot(countries, gdps, color='red', marker='x', label='GDP per capita')
ax2.set_ylabel('GDP per capita', color='red')
ax2.tick_params('y', labelcolor='red')

# Add a colorbar to explain the GDP color mapping in the scatter plot
cbar = fig.colorbar(ax1.collections[0], ax=ax1)  # Associate colorbar with scatter plot
cbar.ax.set_ylabel('GDP per capita (Color)')

plt.title('Score and GDP per capita vs Country', loc='right',
          fontdict={'color': 'blue', 'size': 20, 'style': 'italic'})
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame

# Sort the DataFrame by 'GDP per capita' in descending order and get top 10
top_10_gdp = df.sort_values(by=['GDP per capita'], ascending=False).head(10)

# Extract data for plotting
countries = top_10_gdp['Country or region']
scores = top_10_gdp['Score']
gdps = top_10_gdp['GDP per capita']

# Create the plot
plt.figure(figsize=(10, 6))
plt.bar(countries, scores, color='skyblue', label='Score')  # Bar plot for Score
plt.plot(countries, gdps, color='red', marker='o', label='GDP per capita')  # Line plot for GDP

# Customize the plot
plt.title('Top 10 Countries: GDP per capita vs Score')
plt.xlabel('Country')
plt.ylabel('Score / GDP per capita')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.legend()  # Show legend

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```
